```python
import pandas as pd
import numpy as np
import yfinance as yf
from statsmodels.tsa.arima.model import ARIMA
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Luego debe construirse una función para obtener los datos de Yahoo Finance
def get_stock_data(ticker):
    """
    Función para obtener datos de Yahoo Finance para un símbolo dado.
    """
    try:
        df = yf.download(ticker, start="2010-01-01", end="2023-01-01")
        df = df['Adj Close'].dropna()
        return df
    except Exception as e:
        print(f"No se pudieron obtener los datos para {ticker}: {str(e)}")
        return None
```

```python
# Aprovechando el código podemos proponer una función para entrenar el modelo ARIMA
def train_arima(train):
    """
    Función para entrenar el modelo ARIMA con los datos de entrenamiento dados.
    """
    model = ARIMA(train, order=(5,1,0))
    model_fit = model.fit()
    return model_fit
```

```python
# Hacemos lo propio con una función para entrenar el modelo MLP, este es el modelo perceptrón multicapa
def train_mlp(train, lags=5):
    """
    Función para entrenar el modelo MLP con los datos de entrenamiento dados y un número dado de lags.
    """
    X, y = [], []
    for i in range(lags, len(train)):
        X.append(train[i-lags:i])
        y.append(train[i])
    X, y = np.array(X), np.array(y)
    model = MLPRegressor(hidden_layer_sizes=(100,), max_iter=500)
    model.fit(X, y)
    return model, lags
```

```python
# Ahora establecemos una función para evaluar el modelo ARIMA
def evaluate_arima(model, train, test):
    """
    Función para evaluar el modelo ARIMA en los datos de prueba dados.
    """
    history = list(train)
    predictions = []
    for t in range(len(test)):
        model_fit = ARIMA(history, order=(5,1,0)).fit()
        yhat = model_fit.forecast(steps=1)[0]
        predictions.append(yhat)
        history.append(test[t])
    mse = mean_squared_error(test, predictions)
    return predictions, mse
```

```python
# Hacemos lo propio para una función que evalúe el modelo MLP
def evaluate_mlp(model, test, lags):
    """
    Función para evaluar el modelo MLP en los datos de prueba dados.
    """
    X_test = []
    for i in range(lags, len(test)):
        X_test.append(test[i-lags:i])
    X_test = np.array(X_test)
    predictions = model.predict(X_test)
    return predictions
```

```python
# Una vez tenemos los dos modelos evaluados podemos plantear una función comparativa de los modelos ARIMA y MLP
def compare_models(train, test):
    """
    Función para comparar los modelos ARIMA y MLP en términos de error cuadrático medio (MSE).
    """
    arima_model = train_arima(train)
    mlp_model, lags = train_mlp(train)

    arima_predictions, arima_mse = evaluate_arima(arima_model, train, test)
    mlp_predictions = evaluate_mlp(mlp_model, test, lags)

    results = {
        'ARIMA': arima_mse,
        'MLP': mean_squared_error(test[lags:], mlp_predictions)
    }

    best_model = min(results, key=results.get)
    return best_model, results, arima_predictions, mlp_predictions, lags
```

```python
# Graficamos los resultados
def plot_results(train, test, arima_predictions, mlp_predictions, lags):
    """
    Función para graficar los resultados de ARIMA y MLP junto con los datos de entrenamiento y prueba.
    """
    plt.figure(figsize=(14, 7))
    plt.plot(train.index, train, label='Train Data', color='blue')
    plt.plot(test.index, test, label='Test Data', color='red', alpha=0.5)
    
    # Aseguramos que las predicciones comiencen desde el mismo punto que los datos de prueba
    arima_pred_index = test.index[:len(arima_predictions)]
    mlp_pred_index = test.index[lags:lags+len(mlp_predictions)]
    
    plt.plot(arima_pred_index, arima_predictions, label='ARIMA Predictions', color='green')
    plt.plot(mlp_pred_index, mlp_predictions, label='MLP Predictions', color='orange')
    
    plt.title('ARIMA vs MLP Model - Training and Test Predictions')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```python
# Ahora planteamos la función principal para ejecutar el análisis, este es nuestro bot
def main():
    """
    Función principal para ejecutar el análisis comparativo entre ARIMA y MLP para predicción de series temporales.
    """
    ticker = 'JPM'  # Cambiar el ticker según el símbolo de acción deseado
    df = get_stock_data(ticker)
    
    if df is None:
        return
    
    train_size = int(len(df) * 0.8)
    train, test = df[:train_size], df[train_size:]

    best_model, results, arima_predictions, mlp_predictions, lags = compare_models(train, test)

    print("Best Model: ", best_model)
    print("Results: ", results)

    plot_results(train, test, arima_predictions, mlp_predictions, lags)

if __name__ == "__main__":
    main()
```

```python
# Apliquemos el procceso de automatización
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from scipy.optimize import minimize
from threading import Timer

# Función para obtener datos de múltiples activos
def get_data(tickers, start='2020-01-01', end='2023-01-01'):
    data = yf.download(tickers, start=start, end=end)['Adj Close']
    return data

# Función para calcular la relación Sharpe
def sharpe_ratio(weights, mean_returns, cov_matrix, risk_free_rate=0.01):
    portfolio_return = np.sum(mean_returns * weights) * 252
    portfolio_volatility = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(252)
    return (portfolio_return - risk_free_rate) / portfolio_volatility

# Función para minimizar (maximizar la relación Sharpe)
def negative_sharpe_ratio(weights, mean_returns, cov_matrix, risk_free_rate=0.01):
    return -sharpe_ratio(weights, mean_returns, cov_matrix, risk_free_rate)

# Función para optimizar la cartera
def optimize_portfolio(tickers, data):
    returns = data.pct_change().dropna()
    mean_returns = returns.mean()
    cov_matrix = returns.cov()

    # Verificar que la matriz de covarianza no contenga valores NaN o infinitos
    if np.any(np.isnan(cov_matrix)) or np.any(np.isinf(cov_matrix)):
        raise ValueError("La matriz de covarianza contiene valores no válidos.")
    
    # Restricciones: la suma de los pesos debe ser 1
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bounds = tuple((0, 1) for _ in range(len(tickers)))
    initial_weights = [1/len(tickers)] * len(tickers)
    
    optimal_result = minimize(negative_sharpe_ratio, initial_weights, args=(mean_returns, cov_matrix),
                              method='SLSQP', bounds=bounds, constraints=constraints)
    
    # Normalizar pesos para asegurar que sumen 1
    optimal_weights = optimal_result.x / np.sum(optimal_result.x)
    
    return optimal_weights

# Función para seleccionar los activos más atractivos
def select_top_assets(data, num_assets=5):
    returns = data.pct_change().dropna()
    mean_returns = returns.mean()
    cov_matrix = returns.cov()
    
    sharpe_ratios = {}
    for ticker in data.columns:
        weights = np.zeros(len(data.columns))
        weights[data.columns.get_loc(ticker)] = 1.0
        sharpe_ratios[ticker] = sharpe_ratio(weights, mean_returns, cov_matrix)
    
    sorted_assets = sorted(sharpe_ratios.items(), key=lambda x: x[1], reverse=True)
    top_assets = [asset[0] for asset in sorted_assets[:num_assets]]
    
    return top_assets

# Función principal para actualizar y optimizar la cartera
def update_and_optimize_portfolio():
    tickers = ['SPY', 'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'NVDA', 'JPM', 'V']
    
    try:
        data = get_data(tickers, start='2020-01-01', end='2023-01-01')
        data = data.dropna(axis=1)  # Eliminar columnas con datos faltantes

        if data.empty:
            print("No se pudieron obtener datos válidos para los tickers proporcionados.")
            return

        top_tickers = select_top_assets(data)
        optimal_weights = optimize_portfolio(top_tickers, data[top_tickers])
        
        print("Pesos óptimos de la cartera:")
        for ticker, weight in zip(top_tickers, optimal_weights):
            print(f"{ticker}: {weight:.4f}")
        
        plot_portfolio(top_tickers, optimal_weights, data)
        
    except Exception as e:
        print(f"Error al obtener datos o al optimizar la cartera: {e}")

# Función para visualizar la cartera optimizada
def plot_portfolio(tickers, weights, data):
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.barplot(x=tickers, y=weights, ax=ax)
    ax.set_title('Pesos óptimos de la cartera')
    plt.show()
    
    fig = px.pie(values=weights, names=tickers, title='Distribución de la cartera')
    fig.show()

# Función para programar actualizaciones cada 5 minutos
def schedule_updates(interval=20):
    update_and_optimize_portfolio()
    Timer(interval, schedule_updates, [interval]).start()

# Iniciar el proceso de actualización
schedule_updates()








```

```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from pypfopt import risk_models, expected_returns, EfficientFrontier, objective_functions
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices

# Definir los símbolos de los activos
symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']  
# Descargar datos históricos
data = yf.download(symbols, start="2020-01-01", end="2023-12-31")['Adj Close']

# Calcular los retornos diarios
returns = data.pct_change().dropna()

# Utilizar el enfoque bayesiano para estimar los retornos esperados y la matriz de covarianza
mu = expected_returns.mean_historical_return(data, frequency=252)
S = risk_models.CovarianceShrinkage(data).ledoit_wolf()

# Optimizar la cartera usando el modelo de Markowitz con las estimaciones bayesianas
ef = EfficientFrontier(mu, S)
ef.add_objective(objective_functions.L2_reg, gamma=0.1)  # Agregar regularización L2 para una mayor estabilidad
weights = ef.max_sharpe()

# Limpiar los pesos
cleaned_weights = ef.clean_weights()

# Mostrar los pesos optimizados
print("Pesos optimizados:")
print(cleaned_weights)

# Obtener el rendimiento esperado, la volatilidad y el ratio de Sharpe de la cartera optimizada
expected_annual_return, annual_volatility, sharpe_ratio = ef.portfolio_performance()
print("\nRendimiento esperado anual:", expected_annual_return)
print("Volatilidad anual:", annual_volatility)
print("Ratio de Sharpe:", sharpe_ratio)

# Asignación discreta (por ejemplo, si tienes $10,000 para invertir)
latest_prices = get_latest_prices(data)
da = DiscreteAllocation(cleaned_weights, latest_prices, total_portfolio_value=10000)
allocation, leftover = da.lp_portfolio()
print("\nAsignación discreta de la cartera:")
print(allocation)
print("Fondos restantes después de la asignación:", leftover)
```
